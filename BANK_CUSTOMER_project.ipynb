Question 1:

Print customer Id, customer name and average account_balance maintained by each customer for all  of his/her accounts in the bank.(8 Rows)

```python
# 1) Print customer Id, customer name and average account_balance maintained by each customer for all  of his/her accounts in the bank.(8 Rows)

cursor.execute("""
SELECT
    c.customer_id,
    c.customer_name,
    AVG(a.Balance_Amount) AS average_balance
FROM
    BANK_CUSTOMER c
JOIN
     BANK_ACCOUNT_DETAILS a ON c.customer_id = a.customer_id
GROUP BY
    c.customer_id, c.customer_name
LIMIT 8;
""")
result = cursor.fetchall()
for row in result:
    print(row)


```

Question 2:

Print customer_id , account_number and balance_amount , condition that if balance_amount is nil then assign transaction_amount  for account_type = "Credit Card"(4 Rows)


```python
#Print customer_id , account_number and balance_amount , condition that if balance_amount is nil then assign transaction_amount  for account_type = "Credit Card"(4 Rows)

cursor.execute("""
SELECT
    b.customer_id,
    b.account_number,
    CASE
    WHEN b.Balance_amount IS NULL AND Account_type = 'Credit card' THEN t.transaction_amount
    ELSE b.Balance_amount
    END AS balance_amount
FROM
    BANK_ACCOUNT_DETAILS b
LEFT JOIN
    BANK_ACCOUNT_TRANSACTION t ON b.account_number = t.account_number

LIMIT 4;
""")

result = cursor.fetchall()
for row in result:
    print(row)
```

Question 3:

Print account_number and balance_amount , transaction_amount,Transaction_Date from Bank_Account_Details and bank_account_transaction for all the transactions occurred during march,2020 and april, 2020(12 Rows)


```python
#Print account_number and balance_amount , transaction_amount,Transaction_Date from Bank_Account_Details and bank_account_transaction for all the transactions occurred during march,2020 and april, 2020(12 Rows)

cursor.execute("""
SELECT
    b.account_number,
    b.Balance_amount,
    t.transaction_amount,
    t.transaction_date
FROM
    BANK_ACCOUNT_DETAILS b
JOIN
    BANK_ACCOUNT_TRANSACTION t ON b.account_number = t.account_number
WHERE
    t.transaction_date BETWEEN '2020-03-01' AND '2020-04-30'
LIMIT 12;
""")
result = cursor.fetchall()
for row in result:
    print(row)
```

Question 4:

Print all of the customer id, account number,  balance_amount, transaction_amount , Transaction_Date  from bank_customer, Bank_Account_Details and bank_account_transaction tables where excluding all of their transactions in march, 2020  month (22 Rows)


```python
cursor.execute("""
SELECT
    c.customer_id,
    b.account_number,
    b.balance_amount,
    t.transaction_amount,
    t.transaction_date
FROM
    BANK_CUSTOMER c
JOIN
    BANK_ACCOUNT_DETAILS b ON c.customer_id = b.customer_id
LEFT JOIN
    BANK_ACCOUNT_TRANSACTION t ON b.account_number = t.account_number
WHERE
    NOT(
      strftime('%Y',t.transaction_date) = '2020'
      AND strftime('%m',t.transaction_date)= '03'
    );
""")
result = cursor.fetchall()
for row in result:
    print(row)
```

Question 7:

Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by  subtracting only negative transaction_amounts for Relationship_type ="P" ( P - means  Primary , S - means Secondary ) .(27 Rows)


```python
cursor.execute("""
SELECT
      c.customer_id,
      b.account_number,
      b.account_type,
      b.balance_amount,
      b.relationship_type,
      SUM(CASE
           WHEN t.transaction_amount < 0 THEN t.transaction_amount ELSE 0 END) AS deducted_balance
FROM
    BANK_CUSTOMER c
JOIN
    BANK_ACCOUNT_DETAILS b ON c.customer_id = b.customer_id
LEFT JOIN
    BANK_ACCOUNT_TRANSACTION t ON b.account_number = t.account_number
WHERE b.relationship_type IN ('P', 'S')
GROUP BY
    c.customer_id,
    b.account_number,
    b.account_type,
    b.balance_amount,
    b.relationship_type
""")
result = cursor.fetchall()
for row in result:
    print(row)
```

```python

cursor.execute("""
SELECT DISTINCT
    c.customer_id,
    c.customer_name
FROM
    Bank_Account_Details a
JOIN
    BANK_CUSTOMER c
    ON a.customer_id = c.customer_id
WHERE
    a.customer_id IN (
        SELECT customer_id
        FROM Bank_Account_Details
        WHERE account_type = 'Credit Card'
    )
    AND a.customer_id IN (
        SELECT customer_id
        FROM Bank_Account_Details
        WHERE account_type = 'Add-on Credit Card'
    )
    AND a.account_type = 'SAVINGS';
"""
)
result = cursor.fetchall()
for row in result:
    print(row)
```

```python
cursor.execute("""
SELECT
    c.customer_id,
    c.customer_name
FROM
    Bank_Account_Details a
JOIN
    BANK_CUSTOMER c
    ON a.customer_id = c.customer_id
WHERE
    a.account_type IN ('Credit Card', 'Add-on Credit Card', 'SAVINGS')
GROUP BY
    c.customer_id, c.customer_name
HAVING
    SUM(CASE WHEN a.account_type = 'Credit Card' THEN 1 ELSE 0 END) > 0
    AND SUM(CASE WHEN a.account_type = 'Add-on Credit Card' THEN 1 ELSE 0 END) > 0
    AND SUM(CASE WHEN a.account_type = 'SAVINGS' THEN 1 ELSE 0 END) > 0;
""")
result = cursor.fetchall()
for row in result:
    print(row)
```

Question 11:

Display  records of SAVINGS account linked with Credit cards" account_type and its credit

aggregate sum of transaction amount. (1 Row)

Ref: Check linking relationship in bank_transaction_relationship_details.
        # Check transaction_amount in bank_account_transaction.


```python
cursor.execute("""
SELECT
    s.account_number AS savings_account,
    c.account_number AS credit_card_account,
    SUM(t.transaction_amount) AS total_credit_transactions
FROM
    Bank_Account_Details s
JOIN
    Bank_Account_Relationship_Details r
    ON s.account_number = r.account_number
JOIN
    Bank_Account_Details c
    ON r.linking_account_number = c.account_number
JOIN
    BANK_ACCOUNT_TRANSACTION t
    ON c.account_number = t.account_number
WHERE
    s.account_type = 'SAVINGS'
    AND c.account_type = 'Credit Card';
""")
result = cursor.fetchall()
for row in result:
    print(row)
```
